```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_data = pd.read_csv('Datasets/Rank.csv')
df = file_data.copy()
df = df[( (df['Region'] == 'Kolkata') | (df['Region'] == 'Delhi') | (df['Region'] == 'Chennai') | (df['Region'] == 'Mumbai') ) ]
df = df.reset_index()
df.drop('index',axis='columns',inplace=True)
df.drop('Rank',axis='columns',inplace=True)
df['Rank'] = df.index
df['Rank'] = df.Rank + 1
unique = df['Region'].value_counts()
plt.pie(unique.values,labels=unique.index,autopct = '%.2f')
plt.show()
df.to_excel('Datasets/Zone1.xlsx')
```
